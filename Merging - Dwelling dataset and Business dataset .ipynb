## Need to clean up dataset
The aim of this data set is to extract the number of dwelling in individual clue_small_area for each year.
1. Use data with Status = COMPLETED - To avoild duplicates.
2. Identify required fields 
3. Produce the total Number of residential dwelling for individual clue_small_ares for each year.
4. Identify the new dataset field types and adjust.
5. Check data for missing values

```python
#1. Use data with Status = COMPLETED
data = data.loc[data['status'] == 'COMPLETED']
```

```python
#2. Identify required fields
data.columns
```

```python
data.iloc[:,[3,4,13,14,15,16,17,18]]
```

```python
#3. Produce the total Number of residential dwelling for individual clue_small_ares for each year.
data['Total'] = data.iloc[:,[13,14,15,16,17,18]].sum(axis=1)
```

```python
data.head(3)
```

```python
data= data.iloc[:,[3,4,42]]
```

```python
#. Identify the new dataset field types and adjust.
data.dtypes
```

```python
# Changing year_completed in Year
data['year_completed']=data['year_completed'].astype(int)
```

```python
data.dtypes
```

```python
#5. Check data for missing values
data.isna().sum()
```

```python
# Lets have a view of the data
data = data.groupby(['year_completed','clue_small_area']).agg({'Total':'sum'})
```

```python
data.head(20)
```

```python
data = data.reset_index()
```

## Importing the second dataset - business
1. Check dataset for null values
2. Group data together per year, and individual 
3. Identify unecessary fields
4. Check the data types

```python
databuz = pd.read_csv('Business.csv')
```

```python
databuz.head(5)
```

```python
#1. Check dataset for null values
databuz.isna().sum()
```

```python
#2. Group data together per year, and individual 
dataGroup = databuz.groupby(['census_year','clue_small_area']).sum()
```

```python
dataGroup.head(30)
```

```python
#3. Identify unecessary fields
# The block_id is not required
dataGroup.drop(columns =['block_id'], inplace=True)
```

```python
dataGroup = dataGroup.reset_index()
```

```python
#4. Check the data types
dataGroup.dtypes
```

```python
## Joining the two datasets together
1. Set the year as same column
2. Merge data on year and clue_small_area
```

```python
#1. Set the year as same column
data.dtypes
```

```python
data.rename(columns={'year_completed':'census_year'},inplace=True)
```

```python
data.dtypes
```

```python
#2. Merge data on year and clue_small_area
merged = pd.merge(dataGroup, data, how='left',on=['census_year','clue_small_area'])
```

```python
merged.head(10)
```

```python
merged.to_csv('datamerged.csv',index = False)
```
