# Machine learning linearRegression- Dwelling and industries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('machinedata.csv')
```

```python
data.head(3)
```

```python
# Encoding 'clue_small_area'
le = preprocessing.LabelEncoder()
data['clue_small_area'] = le.fit_transform(data['clue_small_area'].values)
```

```python
data.head(3)
```

```python
data.describe()
```

```python
# All the values are going to be scaled except for target value
warnings.filterwarnings('ignore')
scaler = StandardScaler()
data.iloc[:,2:-2] = scaler.fit_transform(data.iloc[:,2:-2])
```

```python
data.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-2], data.iloc[:,-1], test_size=0.40, random_state=42)
```

```python
# creating model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
import warnings
y_pred = model.predict(X_test)
```

```python
print(model.coef_)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import median_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.linear_model import LogisticRegression
```

```python
def linear_model(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    warnings.filterwarnings('ignore')
    s = model.score(X_test,y_test)
    r = r2_score(y_test,y_pred)
    m = median_absolute_error(y_test, y_pred)
    print("Performance of Linear model")
    print(f"Model score: {s.round(2)}")
    print(f"The r2_score: {r.round(2)}")
    print(f"The mean absolute error: {m.round(2)}")
```

```python
linear_model(X_train, X_test, y_train, y_test)
```

```python
def logistic_model(X_train, X_test, y_train, y_test, penalty, solver):
    model2 = LogisticRegression(penalty =penalty,solver =solver)
    model2.fit(X_train, y_train)
    y_pred = model2.predict(X_test)
    s = model2.score(X_test,y_test)
    r = r2_score(y_test,y_pred)
    m = median_absolute_error(y_test, y_pred)
    print("Performance of Logistic Regression")
    print(f"Penalty: {penalty}, solver: {solver}")
    print(f"Model score: {s.round(2)}")
    print(f"The r2_score: {r.round(2)}")
    print(f"The mean absolute error: {m.round(2)}")
    return r
```

```python
#testing logistic model
r1 = logistic_model(X_train, X_test, y_train, y_test, 'none', 'lbfgs')
```

```python
r2 = logistic_model(X_train, X_test, y_train, y_test, 'l2', 'lbfgs')
```

```python
r3 = logistic_model(X_train, X_test, y_train, y_test, 'l1', 'liblinear')
```

```python
r4 = logistic_model(X_train, X_test, y_train, y_test, 'l2', 'liblinear')
```

```python
r4 = logistic_model(X_train, X_test, y_train, y_test, 'l2', 'newton-cholesky')
```

## Regression with decision trees

```python
from sklearn.tree import DecisionTreeRegressor  
```

```python
def tree_model(X_train, X_test, y_train, y_test, random, max_depth):
    model = DecisionTreeRegressor(random_state = random,max_depth =max_depth)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    s = model.score(X_test,y_test)
    return s.round(2)
    
```

```python
tree_model(X_train, X_test, y_train, y_test, 10,10)
```

```python
score_3 =[]
for i in range(0,20):
    s =tree_model(X_train, X_test, y_train, y_test, i, 3)
    score_3.append(s)
```

```python
print(max(score_3))
```

```python
plt.plot(score_3)
```

```python
score_4 =[]
for i in range(0,20):
    s =tree_model(X_train, X_test, y_train, y_test, i, 4)
    score_4.append(s)
print(max(score_4))
plt.plot(score_4)
```

```python
score_6 =[]
for i in range(0,20):
    s =tree_model(X_train, X_test, y_train, y_test, i, 6)
    score_6.append(s)
print(max(score_6))
plt.plot(score_6)
```

## Support vector regression

```python
from sklearn.svm import SVR
```

```python
def svr_model(X_train, X_test, y_train, y_test, kernel, C ):
    model = SVR(kernel = kernel, C=C)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    s = model.score(X_test,y_test)
    return s.round(2)
```

```python
score = svr_model(X_train, X_test, y_train, y_test, 'poly',10.0)
print(score)
```

```python
score = svr_model(X_train, X_test, y_train, y_test, 'poly',1.0)
print(score)
```

```python
score = svr_model(X_train, X_test, y_train, y_test, 'linear',10.0)
print(score)
```

```python
score1=[]
for i in range(1,100):
    i= float(i)
    s =svr_model(X_train, X_test, y_train, y_test, 'linear',i)
    score1.append(s)
print(max(score1))
plt.plot(score1)
```
